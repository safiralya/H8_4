# Probability Distributions

**Random Variable** adalah variabel yang kemungkinan nilainya merupakan hasil numerik dari fenomena acak. Ada dua jenis variabel acak yaitu :
1. Discrete: Semua kemungkinan outcomes dapat dihitung (countable) atau memiliki batasan (finite)
2. Continuous : Semua kemungkinan outcomes tidak dapat dihitung (uncountable), umumnya direpresentasikan dengan nilai interval.

## Uniform Distribution

adalah bentuk distribusi probabilitas di mana suatu peristiwa memiliki kemungkinan yang sama untuk terjadi dalam interval tertentu. Biasanya digunakan untuk :
1. menebak ulang tahun
2. Melempar dadu
3. lempar coin
4. spining spinner

```python
from scipy.stats import uniform
```

```python
n = 10000
start = 10
width = 20
data_uniform = uniform.rvs(size=n, loc = start, scale=width)
print(data_uniform)
```

```python
ax = sns.distplot(data_uniform,
                  bins=100,
                  kde=True,
                  color='skyblue',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Uniform Distribution ', ylabel='Frequency')
```

# Hypothesis Testing

Hypothesis testing dalam statistik adalah cara kita menguji hasil survei atau eksperimen untuk melihat apakah kita memiliki hasil yang bermakna. Hal yang perlu dilakukan adalah :
1. Figure out your null hypothesis (fakta yang diterima),
2. State your null hypothesis,
3. Choose what kind of test you need to perform,
4. Either support or reject the null hypothesis.

* If p-value > alpha: Fail to reject the null hypothesis (i.e. not signifiant result).
* If p-value <= alpha: Reject the null hypothesis (i.e. significant result).

## One Population Proportion

**Research Question**

Pada tahun-tahun sebelumnya, 52% orang tua percaya bahwa elektronik dan media sosial adalah penyebab kurang tidur remaja mereka. Apakah saat ini semakin banyak orang tua yang percaya bahwa kurang tidur remaja mereka disebabkan oleh elektronik dan media sosial?

Population: Orang tua dengan remaja (usia 13-18) Parameter of Interest: p

Null Hypothesis: p = 0.52 Alternative Hypthosis: p > 0.52 (note that this is a one-sided test)

Data: 1018 orang disurvei. 56% dari mereka yang disurvei percaya bahwa kurang tidur remaja mereka disebabkan oleh elektronik dan media sosial.

```python
n = 1018
pnull = .52
phat = .56
```

```python
sm.stats.proportions_ztest(phat * n, n, pnull, alternative='larger')
```

**Conclusion of the hypothesis test**

Karena nilai p-value dari z-test cukup kecil, kita dapat reject the Null hypothesis bahwa persentase orang tua yang percaya bahwa kurang tidur remaja mereka disebabkan oleh elektronik dan media sosial sama dengan tahun-tahun sebelumnya. Perkiraan yaitu 52%.

## Difference in Population Means

**Research Question**

Dengan mempertimbangkan data NHANES, apakah pria memiliki Body Mass Index rata-rata yang lebih tinggi secara signifikan daripada wanita?

Population: Adults in the NHANES data. Parameter of Interest: 12, Body Mass Index.

Null Hypothesis: 1=2 Alternative Hypthosis: 12

Data:

2976 Females

1=29.94

1=7.75

2759 Male Adults

2=28.78

2=6.25

12=1.16

```python
url = "https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/nhanes_2015_2016.csv"
da = pd.read_csv(url)
da.head()
```

```python
females = da[da["RIAGENDR"] == 2]
male = da[da["RIAGENDR"] == 1]
```

```python
n1 = len(females)
mu1 = females["BMXBMI"].mean()
sd1 = females["BMXBMI"].std()

(n1, mu1, sd1)
```

```python
n2 = len(male)
mu2 = male["BMXBMI"].mean()
sd2 = male["BMXBMI"].std()

(n2, mu2, sd2)
```

```python
sm.stats.ztest(females["BMXBMI"].dropna(), male["BMXBMI"].dropna(),alternative='two-sided')
```

**Conclusion of the hypothesis test**

Karena p-value (6.59e-10) sangat kecil, kita dapat reject the Null hypothesis bahwa rata-rata BMI laki-laki sama dengan perempuan. Catatan, kita menggunakan alternative="two-sided" dalam z-test karena di sini kita memeriksa ketidaksamaan.
